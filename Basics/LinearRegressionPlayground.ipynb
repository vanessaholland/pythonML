## Fit a line to a set of data to use for future predictions

```python
%matplotlib inline
from pylab import *
import numpy as np

purchaseAmounts = np.random.normal(100.0, 10.0, 1000)
salaries = 10000 + (purchaseAmounts + np.random.normal(1000, 10.0, 1000) ) * 2
scatter(purchaseAmounts, salaries)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(purchaseAmounts, salaries)
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitline = predict(purchaseAmounts)
plt.scatter(purchaseAmounts, salaries)
plt.plot(purchaseAmounts, fitline, c='r')
plt.show()
```

```python
r_value ** 2
# not a good fit obvi, increase that 
```

```python
purchaseAmounts = np.random.normal(100.0, 50.0, 1000)
salaries = 10000 + (purchaseAmounts + np.random.normal(1000, 10.0, 1000) ) * 2
scatter(purchaseAmounts, salaries)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(purchaseAmounts, salaries)
fitline = predict(purchaseAmounts)
plt.scatter(purchaseAmounts, salaries)
plt.plot(purchaseAmounts, fitline, c='y')
plt.show()
```

```python
r_value ** 2
# much better
```

## poly regression

```python
np.random.seed(0)
purchaseAmounts = np.random.normal(100.0, 5.0, 1000)
salaries = np.random.normal(50000, 1000.0, 1000)/purchaseAmounts
scatter(purchaseAmounts, salaries)
```

```python
x = np.array(purchaseAmounts)
y = np.array(salaries)
p3 = np.poly1d(np.polyfit(x, y, 4))

xp = np.linspace(115, 85, 1000)
plt.scatter(x, y)
plt.plot(xp, p3(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y, p3(x))
print(r2)
```
