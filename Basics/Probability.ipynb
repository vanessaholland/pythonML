```python
## Probability Density Function
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
# generate normal (Gaussian)
x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

```python
from scipy.stats import binom, poisson
#generate an example binomial PMF with shape of 10, .5
n, p = 10, .5
data_set = np.arange(0, 10, .001)
plt.plot(data_set, binom.pmf(data_set, n, p))
```

```python
## Probability Mass Function 
using heart attack data
```

```python
heart_data = pd.read_csv('../datasets/heart.csv')
exercise_enduced_angina = heart_data[heart_data["exng"] > 0]
ages_exng = exercise_enduced_angina["age"]
mu = ages_exng.mean()
mu
```

```python
plt.plot(ages_exng, poisson.pmf(ages_exng, mu))
```

```python
## Moments
```

```python
### First Moment (mean)
np.mean(ages_exng)
```

```python
### Second Moment (variance)
np.var(ages_exng)
```

```python
### Third Moment (skew, negative left, positive right)
from scipy.stats import skew
skew(ages_exng)
```

```python
### Fourth Moment (kurtosis, shape of the tail)
from scipy.stats import kurtosis
kurtosis(ages_exng)
```
